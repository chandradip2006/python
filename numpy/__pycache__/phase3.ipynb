# advanced operation with business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
print("====== Zomato sales analysis ======")
print("sales data shape: ", sales_data.shape)
print("sales adata for first 3 restaurants: \n", sales_data[:3])
# print("sales adata for first 3 restaurants: \n", sales_data[: , 1:])
```

**total sales per year**

```python
print(np.sum(sales_data , axis=0))
yearly_total=np.sum(sales_data[:,1:] , axis=0)
print("yearly total sales: \n", yearly_total)
```

**minimum sales per restaurant**

```python
min_sales=np.min(sales_data[:,1:] , axis=1)
print("min sales: \n", min_sales)
```

**maximum sales per year**

```python
max_sales=np.max(sales_data[:,1:] , axis=0)
print("max sales per year: \n", max_sales)
```

**average sales per restaurant**

```python
average_sales=np.mean(sales_data[:,1:] , axis=1)
print("average sales per restaurant: \n", average_sales)
```

**cumulative sum per restaurant**

```python
cumsum_sales=np.cumsum(sales_data[:,1:] , axis=1)
print("cumulative sum of sales: \n", cumsum_sales)
# Plotting the sales data
# years = ['2021', '2022', '2023', '2024']
# # Create a bar plot for each restaurant
# for i in range(0, 5):
#     plt.plot(years, sales_data[i, 1:], label=f'Restaurant {int(sales_data[i, 0])}')
# plt.xlabel('Years')

# plt.ylabel('Sales (in $)')
# plt.title('Restaurant Sales Over Years')
# plt.legend()
# plt.show()

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum_sales, axis=0))
plt.xlabel('Years')
plt.ylabel('Average Cumulative Sales (in $)')
plt.title('Average Cumulative Sales Over Years across all restaurants')
plt.grid(True)
plt.show()
```

# vector operations

```python
vector1=np.array([1 , 2 , 3 , 4 , 5])
vector2=np.array([6 , 7 , 8 , 9 , 10])
print("vector addition: \n" , vector1+vector2)
print("vector multiplication: \n" , vector1*vector2)
print("vector dot product: \n" , np.dot(vector1,vector2)) 
angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2))) 
print("angle between vector1 and vector2: \n" , angle)            
```

```python
restau_types = ['Paradise Biryani', 'Beijing Bites', 'Pizza Hub', 'Burger Point', 'Chai Point']
vectorized_upper = np.vectorize(str.upper)
print("vectorized upper: \n", vectorized_upper(restau_types))
```
