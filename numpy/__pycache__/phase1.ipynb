# list vs numpy array

```python
py_list=[1 , 2,3 ]
print("py_list:" , py_list*2)
np_array=np.array(py_list) #element-wise operation
print("np_array:" , np_array*2)
```

# time in numpy array vs python list

```python
import time
start = time.time()
list1 = [i*2 for i in range(1000000)]
print("time taken to create list1:", time.time()-start)
start = time.time()
arr1 = np.arange(1000000)*2
print("time taken to create numpy array:", time.time()-start)
```

# creation of array from scratch

```python
zeros=np.zeros((3, 4))
print("zeros:\n" , zeros)
ones=np.ones((3, 4))
print("ones:\n" , ones)

full=np.full((3, 4), 5)
print("full:\n" , full)

random=np.random.random((3, 4))
print("random:\n" , random)

sequence=np.arange(0 , 20 , 2)
print("sequence:\n" , sequence)
```

# vector , matrix and tensor

```python
vector=np.array([1, 2, 3])
print("vector:\n" , vector)

matrix=np.array([[1, 2, 3], [4, 5, 6]])
print("matrix:\n" , matrix)

tensor=np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("tensor:\n" , tensor)
```

# array properties

```python
arr=np.array([[1, 2, 3], [4, 5, 6]])
print("shape of arr:", arr.shape)
print("size of arr:", arr.size)
print("dimension of arr:", arr.ndim)
print("dtype of arr:", arr.dtype)
```

# array reshape

```python
arr=np.arange(10)
print("arr:\n" , arr)
arr1=arr.reshape(2, 5)
print("reshaped arr:\n" , arr1)

flat=arr1.flatten()
print("flattened arr:\n" , flat)


print("reshaped arr:\n" , arr1)

raveled=arr1.ravel()
print("raveled arr:\n" , raveled)

print("reshaped arr:\n" , arr1)

transposed=arr1.T
print("transposed arr:\n" , transposed)

flat1=flat.reshape(1 , 10)
# print("reshaped flat:\n" , flat.reshape(1 , 10))
trans1=flat1.T
print("transposed flat:\n" , trans1)
```
